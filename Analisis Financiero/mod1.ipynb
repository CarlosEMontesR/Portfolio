```python
fb['Price1'] = fb['Close'].shift(-1) #Precio de cierre anterior
```

```python
fb.head()
```

```python
fb['PriceDiff'] = fb['Price1'] - fb['Close']
```

```python
fb.head(10)
```

```python
fb['Return'] = fb['PriceDiff']/fb['Close']
```

```python
fb.head()
```

```python
fb['Direction'] = [1 if fb.loc[ei, 'PriceDiff'] > 0 else -1 for ei in fb.index]
```

```python
fb.head()
```

```python
#Media Movil
fb['Average3'] = (fb['Close'] + fb['Close'].shift(1) + fb['Close'].shift(2))/3
```

```python
fb.head()
```

```python
#Media movil sobre cualquier numero de dias
fb['MA40'] = fb['Close'].rolling(40).mean() #40 DÃ­as
fb['MA200'] = fb['Close'].rolling(200).mean() #200 DÃ­as
```

```python
fb['Close'].plot()
fb['MA40'].plot()
fb['MA200'].plot()
```

```python
ms = pd.DataFrame(pd.read_csv('MSFT.csv'))
```

```python
ms.head()
```

```python
ms['MA10'] = ms['Close'].rolling(10).mean() #SeÃ±al rÃ¡pida
ms['MA50'] = ms['Close'].rolling(50).mean() #SeÃ±al lenta
```

```python
ms['Close'].plot(legend=True)
ms['MA10'].plot(legend=True)
ms['MA50'].plot(legend=True)
```

```python
ms['Shares'] = [1 if ms.loc[ei, 'MA10']>ms.loc[ei, 'MA50'] else 0 for ei in ms.index]
```

```python
ms
```

```python
ms['Close1'] = ms['Close'].shift(-1)
ms.iloc[500:505,:]
```

```python
ms['Profit'] = [ms.loc[ei, 'Close1']-ms.loc[ei,'Close'] if ms.loc[ei, 'Shares']==1 else 0 for ei in ms.index]
```

```python
ms['Profit'].plot()
```

```python
ms['Wealth'] = ms['Profit'].cumsum()
```
