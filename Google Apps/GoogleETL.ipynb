```python
import pandas as pd 
import numpy as np
import matplotlib.pyplot as plt
```

```python
csv = pd.read_csv("googleplaystore.csv")
# Cargamos el archivo que se va a tratar
```

```python
csv.shape
# Verificamos las dimensiones de nuestro csv
```

```python
nuevas_columnas = ["Applicacion","Categoria", "Calificacion", "Comentarios", "Peso", "Instalaciones", "Tipo", "Precio",
                   "Publico objetivo", "Genero", "Ultima Actualizacion", "Version Actual", "Version de Andriod"]
# Se crea una variable que contiene los nuevos nombres de las columnas
```

```python
df = pd.DataFrame(csv)
# Convertimos nuestro archivo CSV a un DataFrame
```

```python
df.columns = nuevas_columnas
# Asignamos un nuevo nombre a las columnas
```

```python
df
```

```python
lista_df = [df["Applicacion"],df["Categoria"], df["Calificacion"], df["Comentarios"], df["Peso"], df["Instalaciones"],
             df["Tipo"], df["Precio"], df["Publico objetivo"], df["Genero"], df["Ultima Actualizacion"], 
             df["Version Actual"], df["Version de Andriod"]]
```

```python
c = 0
for i in lista_df:
    valor = i.isnull().sum()
    name = df.columns[c]
    c += 1
 
    print("La cantidad de valores nulos es de {} hallazgos en la columna {}".format(valor, name))
# Escribimos un breve ciclo for para determinar cuantos valores nulos hay en cada columna
```

```python
df_sin_na = df.dropna()
# Eliminamos los valores nulos para poder proceder analizar el dataset
df_sin_na.shape
# verificamos el nuevo tamaño de nuestro dataset
```

```python
df_sin_na
```

```python
appsxcategoria = df_sin_na['Categoria'].value_counts()
diez_prim = pd.DataFrame(appsxcategoria.head(10))
diez_prim


# ¿Cuáles son las 10 categorias con mayor cantidad de aplicaciones?
```
