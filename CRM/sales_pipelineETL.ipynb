```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv("sales_pipeline.csv")
df
```

```python
nombres = ["ID_Oportunidad", "Agente_Ventas", "Producto", "Cuenta","Etapa_Operacion", "Fecha_Apertura",
           "Fecha_Cierre", "Valor_Cierre"]
```

```python
df.columns = nombres
```

```python
columnas = [df["ID_Oportunidad"], df["Agente_Ventas"], df["Producto"], df["Etapa_Operacion"], 
            df["Fecha_Apertura"], df["Fecha_Cierre"], df["Valor_Cierre"]]
```

```python
df
```

```python
c = 0
for i in columnas:
    valor = i.isnull().sum() # Nos permite sumar el numero de valores nulos presentes en la categoria
    name = df.columns[c] # Nos permite accder 1 a 1 a las columnas
    c += 1 # Contador que nos permite iterar entre columnas
 
    print("La cantidad de valores nulos es de {} hallazgos en la columna {}".format(valor, name))
```

```python
df_nona = df.dropna()
df_nona
```

```python
columnas_nona = [df_nona["ID_Oportunidad"], df_nona["Agente_Ventas"], df_nona["Producto"], df_nona["Etapa_Operacion"], 
            df_nona["Fecha_Apertura"], df_nona["Fecha_Cierre"], df_nona["Valor_Cierre"]]
```

```python
c = 0
for i in columnas_nona:
    valor = i.isnull().sum() # Nos permite sumar el numero de valores nulos presentes en la categoria
    name = df_nona.columns[c] # Nos permite accder 1 a 1 a las columnas
    c += 1 # Contador que nos permite iterar entre columnas
 
    print("La cantidad de valores nulos es de {} hallazgos en la columna {}".format(valor, name))
```
