```python
import pandas as pd
import numpy as np
```

```python
accounts = pd.read_csv("accounts.csv")
accounts
```

```python
nuevas_columnas = ["Cuenta", "Sector", "Anio_Establecido", "Ingresos", "Empleados", "Localizacion_Oficina",
                   "Subsidiado"]
# Definimos una lista con un conjunto de nombres para las columnas
```

```python
columnas = [accounts['Cuenta'], accounts['Sector'], accounts['Anio_Establecido'], accounts['Ingresos'], 
            accounts['Empleados'], accounts['Localizacion_Oficina'], accounts['Subsidiado']]
# Defiminos una lista con la llamada de cada una de las columnas para luego aplicarle funciones
```

```python
c = 0
for i in columnas:
    valor = i.isnull().sum() # Nos permite sumar el numero de valores nulos presentes en la categoria
    name = accounts.columns[c] # Nos permite accder 1 a 1 a las columnas
    c += 1 # Contador que nos permite iterar entre columnas
 
    print("La cantidad de valores nulos es de {} hallazgos en la columna {}".format(valor, name))
```

```python
accounts.columns = nuevas_columnas
del accounts['Subsidiado'] # Se elimina la columna de subsidiado por tener un 90% de NaN
accounts
```

```python
sectores = accounts['Sector'].value_counts()
sectores
```

```python
sectores = accounts['Localizacion_Oficina'].value_counts()
sectores
```
